```python
#Storing new names in a variable
new_column_name ={
    "show_id" : "SHOW_ID",
    "type" : "SHOW_TYPE",
    "title" : "SHOW_TITLE",
    "director" : "SHOW_DIRECTOR",
    "cast" : "SHOW_CAST" ,
    "country" : "SHOW_COUNTRY",
    "date_added" : "SHOW_ADDED_DATE",
    "release_year" : "SHOW_RELEASE_YEAR",
    "rating" : "SHOW_RATING",
    "duration" : "SHOW_DURATION",
    "listed_in" : "SHOW_LISTED_IN",
    "description":"SHOW_DESCRIPTION"
    }

# rename the old columns with new names
new_data = data_val.rename(columns=new_column_name)
```

```python
# Data Visualization
cat_cols=new_data.select_dtypes(include=['object']).columns
num_cols = new_data.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(cat_cols)
print("Numerical Variables:")
print(num_cols)

for col in num_cols:
    print(col)
    print('Skew :', round(new_data[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    new_data[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=new_data[col])
    plt.show()
```
